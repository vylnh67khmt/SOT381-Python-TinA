```python
w=float(input("Nhập vào độ dài cạnh thứ nhất: "))
h=float(input("Nhập vào độ dài cạnh thứ hai:"))
while w>0.0 and h<=100.0:
  chuvi=(w+h)*2
  dientich=w*h
  print(f"Chu vi hình chữ nhật là: {chuvi:.2f}")
  print(f"Diện tích hình chữ nhật là: {dientich:.2f}")
  break
```
