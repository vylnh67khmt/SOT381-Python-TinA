```python
w=float(input("Nhập vào độ dài cạnh thứ nhất: "))
while True:
  if w>0.0 and w<=100.0:
    break
else:
  print(f"Mời bạn nhập lại giá trị")
h=float(input("Nhập vào độ dài cạnh thứ hai:"))
while True:
  if h>0.0 and h<=100.0:
    break
  else:
    print("Mời bạn nhập lại giá trị")
chuvi=(w+h)*2
dientich=w*h
print(f"Chu vi hình chữ nhật là: {chuvi:.2f}")
print(f"Diện tích hình chữ nhật là: {dientich:.2f}")
```

```python
a=float(input("Nhập vào số thứ nhất: "))
b=float(input("Nhập vào số thứ hai: "))
c=float(input("Nhập vào số thứ ba: "))
max=a
if b>max:
  max=b
  print(f"Số lớn nhất là: {max:.1f}")
if c>max:
  max=c
  print(f"Số lớn nhất là: {max:.1f}")
else:
  print(f"Số lớn nhất là: {max:.1f}")
```
